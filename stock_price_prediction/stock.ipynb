```python
"""
Multi-Stock Risk-Return Analysis System
여러 종목의 리스크와 수익률을 분석하는 시스템
"""

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

class StockRiskReturnAnalyzer:
    """주식 리스크-수익률 분석기"""
    
    def __init__(self, tickers, start_date=None, end_date=None):
        """
        Parameters:
        -----------
        tickers : list
            분석할 종목 리스트 (예: ['AAPL', 'MSFT', 'TSLA'])
        start_date : str
            시작일 (YYYY-MM-DD 형식, 기본값: 3년 전)
        end_date : str
            종료일 (YYYY-MM-DD 형식, 기본값: 오늘)
        """
        self.tickers = tickers
        self.end_date = end_date or datetime.now().strftime('%Y-%m-%d')
        self.start_date = start_date or (datetime.now() - timedelta(days=3*365)).strftime('%Y-%m-%d')
        
        self.prices = None
        self.returns = None
        self.stats = None
        
    def fetch_data(self):
        """yfinance로 주식 데이터 수집"""
        print(f"📊 데이터 수집 중: {', '.join(self.tickers)}")
        print(f"기간: {self.start_date} ~ {self.end_date}\n")
        
        try:
            # 여러 종목 동시 다운로드
            data = yf.download(
                self.tickers,
                start=self.start_date,
                end=self.end_date,
                progress=False
            )
            
            # 종가만 추출
            if len(self.tickers) == 1:
                self.prices = data['Close'].to_frame()
                self.prices.columns = self.tickers
            else:
                self.prices = data['Close']
            
            # 결측치 제거
            self.prices = self.prices.dropna()
            
            print(f"✅ 데이터 수집 완료: {len(self.prices)} 거래일\n")
            return self.prices
            
        except Exception as e:
            print(f"❌ 데이터 수집 실패: {e}")
            return None
    
    def calculate_returns(self):
        """일별 수익률 계산"""
        if self.prices is None:
            print("⚠️ 먼저 fetch_data()를 실행하세요")
            return None
        
        # 일별 수익률 = (오늘 가격 / 어제 가격) - 1
        self.returns = self.prices.pct_change().dropna()
        
        print(f"✅ 수익률 계산 완료: {len(self.returns)} 일\n")
        return self.returns
    
    def calculate_statistics(self, annual_factor=252):
        """종목별 통계 계산
        
        Parameters:
        -----------
        annual_factor : int
            연간 환산 계수 (기본값: 252 거래일)
        """
        if self.returns is None:
            print("⚠️ 먼저 calculate_returns()를 실행하세요")
            return None
        
        stats = pd.DataFrame({
            '평균 일일 수익률 (%)': self.returns.mean() * 100,
            '연간 기대 수익률 (%)': self.returns.mean() * annual_factor * 100,
            '일일 변동성 (%)': self.returns.std() * 100,
            '연간 변동성 (%)': self.returns.std() * np.sqrt(annual_factor) * 100,
            '샤프 비율': (self.returns.mean() / self.returns.std()) * np.sqrt(annual_factor),
            '최대 일일 수익률 (%)': self.returns.max() * 100,
            '최소 일일 수익률 (%)': self.returns.min() * 100
        })
        
        # 소수점 정리
        stats = stats.round(2)
        self.stats = stats
        
        print("=" * 80)
        print("📈 종목별 리스크-수익률 통계")
        print("=" * 80)
        print(stats)
        print("\n")
        
        return stats
    
    def plot_risk_return(self, figsize=(10, 6)):
        """리스크-수익률 산점도"""
        if self.stats is None:
            print("⚠️ 먼저 calculate_statistics()를 실행하세요")
            return
        
        plt.figure(figsize=figsize)
        
        # 산점도
        plt.scatter(
            self.stats['연간 변동성 (%)'],
            self.stats['연간 기대 수익률 (%)'],
            s=200,
            alpha=0.6,
            c=range(len(self.stats)),
            cmap='viridis'
        )
        
        # 종목명 표시
        for ticker, row in self.stats.iterrows():
            plt.annotate(
                ticker,
                (row['연간 변동성 (%)'], row['연간 기대 수익률 (%)']),
                fontsize=11,
                ha='center',
                va='bottom'
            )
        
        plt.xlabel('연간 변동성 (Risk) %', fontsize=12)
        plt.ylabel('연간 기대 수익률 (Return) %', fontsize=12)
        plt.title('Risk-Return Analysis (리스크-수익률 분석)', fontsize=14, fontweight='bold')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    def plot_correlation_matrix(self, figsize=(10, 8)):
        """상관관계 행렬 히트맵"""
        if self.returns is None:
            print("⚠️ 먼저 calculate_returns()를 실행하세요")
            return
        
        corr_matrix = self.returns.corr()
        
        plt.figure(figsize=figsize)
        sns.heatmap(
            corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={'label': 'Correlation'}
        )
        plt.title('종목 간 상관관계 행렬 (Correlation Matrix)', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        print("\n📊 상관관계 해석:")
        print("  1에 가까울수록: 같은 방향으로 움직임")
        print("  0에 가까울수록: 독립적 움직임")
        print(" -1에 가까울수록: 반대 방향으로 움직임\n")
    
    def plot_covariance_matrix(self, figsize=(10, 8)):
        """공분산 행렬 히트맵"""
        if self.returns is None:
            print("⚠️ 먼저 calculate_returns()를 실행하세요")
            return
        
        cov_matrix = self.returns.cov() * 252  # 연간 환산
        
        plt.figure(figsize=figsize)
        sns.heatmap(
            cov_matrix,
            annot=True,
            fmt='.4f',
            cmap='RdYlGn_r',
            square=True,
            linewidths=1,
            cbar_kws={'label': 'Covariance'}
        )
        plt.title('종목 간 공분산 행렬 (Covariance Matrix - 연간환산)', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        print("\n📊 공분산 해석:")
        print("  양수: 함께 상승/하락 경향")
        print("  음수: 반대로 움직이는 경향")
        print("  절대값이 클수록 관계가 강함\n")
    
    def plot_cumulative_returns(self, figsize=(12, 6)):
        """누적 수익률 비교 차트"""
        if self.returns is None:
            print("⚠️ 먼저 calculate_returns()를 실행하세요")
            return
        
        # 누적 수익률 계산 (초기 투자금 $100 가정)
        cumulative = (1 + self.returns).cumprod() * 100
        
        plt.figure(figsize=figsize)
        for ticker in cumulative.columns:
            plt.plot(cumulative.index, cumulative[ticker], label=ticker, linewidth=2)
        
        plt.xlabel('날짜', fontsize=12)
        plt.ylabel('누적 수익 ($100 투자 시)', fontsize=12)
        plt.title('종목별 누적 수익률 비교 (Cumulative Returns)', fontsize=14, fontweight='bold')
        plt.legend(loc='best')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    def generate_report(self):
        """종합 분석 리포트 생성"""
        print("\n" + "="*80)
        print("📋 종합 분석 리포트")
        print("="*80 + "\n")
        
        if self.stats is not None:
            # 최고/최저 종목
            best_return = self.stats['연간 기대 수익률 (%)'].idxmax()
            worst_return = self.stats['연간 기대 수익률 (%)'].idxmin()
            lowest_risk = self.stats['연간 변동성 (%)'].idxmin()
            highest_risk = self.stats['연간 변동성 (%)'].idxmax()
            best_sharpe = self.stats['샤프 비율'].idxmax()
            
            print(f"🏆 최고 기대수익률: {best_return} ({self.stats.loc[best_return, '연간 기대 수익률 (%)']:.2f}%)")
            print(f"📉 최저 기대수익률: {worst_return} ({self.stats.loc[worst_return, '연간 기대 수익률 (%)']:.2f}%)")
            print(f"🛡️ 최저 리스크: {lowest_risk} ({self.stats.loc[lowest_risk, '연간 변동성 (%)']:.2f}%)")
            print(f"⚡ 최고 리스크: {highest_risk} ({self.stats.loc[highest_risk, '연간 변동성 (%)']:.2f}%)")
            print(f"⭐ 최고 샤프비율: {best_sharpe} ({self.stats.loc[best_sharpe, '샤프 비율']:.2f})")
            
            print("\n💡 샤프 비율 = 위험 대비 수익 효율성 (높을수록 좋음)")
        
        print("\n" + "="*80 + "\n")


# ============================================================================
# 실행 예시
# ============================================================================

if __name__ == "__main__":
    # 분석할 종목 리스트
    tickers = ['AAPL', 'MSFT', 'TSLA', 'NVDA', 'GOOGL', 'AMZN', 'META', 'SPY']
    
    # 분석기 생성
    analyzer = StockRiskReturnAnalyzer(tickers)
    
    # 1단계: 데이터 수집
    analyzer.fetch_data()
    
    # 2단계: 수익률 계산
    analyzer.fetch_data()
    analyzer.calculate_returns()
    
    # 3단계: 통계 계산 및 출력
    analyzer.calculate_statistics()
    
    # 4단계: 시각화
    analyzer.plot_risk_return()
    analyzer.plot_correlation_matrix()
    analyzer.plot_covariance_matrix()
    analyzer.plot_cumulative_returns()
    
    # 5단계: 종합 리포트
    analyzer.generate_report()
```
