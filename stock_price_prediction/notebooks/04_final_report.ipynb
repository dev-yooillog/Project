```python
import pandas as pd
import glob
import os

# Load metrics
files = glob.glob("../data/raw/stock_prices_*.csv")
latest_file = max(files, key=os.path.getctime)

prices = pd.read_csv(latest_file, index_col=0, parse_dates=True)
returns = prices.pct_change().dropna()

TRADING_DAYS = 252
RISK_FREE_RATE = 0.02

summary = pd.DataFrame({
    "annual_return": returns.mean() * TRADING_DAYS,
    "annual_volatility": returns.std() * (TRADING_DAYS ** 0.5),
})

summary["sharpe_ratio"] = (
    (summary["annual_return"] - RISK_FREE_RATE) /
    summary["annual_volatility"]
)

summary
```

```python
# Key insights
best_sharpe = summary["sharpe_ratio"].idxmax()
lowest_risk = summary["annual_volatility"].idxmin()
highest_return = summary["annual_return"].idxmax()

best_sharpe, lowest_risk, highest_return
```
